```python
import networkx as nx
import matplotlib.pyplot as plt

# Defina os vértices e arestas do grafo
vertices = [1, 2, 3, 4, 5]
arestas = [(1, 2), (1, 3), (2, 3), (2, 4), (3, 4), (3, 5), (4, 5)]

# Crie o grafo
G = nx.Graph()
G.add_nodes_from(vertices)
G.add_edges_from(arestas)

# Verifique se o grafo é planar usando a fórmula de Euler
V = len(vertices)
E = len(arestas)
F = 2  # Assumindo que o grafo é conectado
if V - E + F == 2:
    print("O grafo é planar.")
else:
    print("O grafo não é planar.")

# Plote o grafo
pos = nx.spring_layout(G)  # Define a posição dos vértices para o layout do Spring
nx.draw(G, pos, with_labels=True, node_size=500, node_color="skyblue", font_size=10)
plt.title("Grafo")
plt.show()
```
