```python
import networkx as nx
import matplotlib.pyplot as plt

# Construa o grafo G
G = nx.Graph()
vertices = [1, 2, 3, 4, 5, 6, 7, 8]
arestas = [(1, 2), (1, 3), (1, 5), (2, 4), (2, 6), (3, 4), (3, 7), (4, 8), (5, 6), (5, 7), (6, 8), (7, 8)]
G.add_nodes_from(vertices)
G.add_edges_from(arestas)

# Verifique se o grafo Ã© planar
if nx.check_planarity(G)[0]:
    print("O grafo Ã© planar.")
else:
    print("O grafo nÃ£o Ã© planar.")

# Plote o grafo
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=300, node_color='lightblue', font_size=10)
plt.title("Grafo")
plt.show()
```
