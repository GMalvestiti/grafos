```python
import networkx as nx
import matplotlib.pyplot as plt

# Defina o grafo G1
G1 = nx.Graph()
G1.add_edges_from([(1, 2), (1, 3), (2, 4), (3, 4), (3, 5), (4, 5)])

# Defina o grafo G2
G2 = nx.Graph()
G2.add_edges_from([(1, 2), (1, 3), (2, 4), (3, 4)])

# Funo para verificar se um grafo  euleriano
def is_eulerian(graph):
    return nx.is_eulerian(graph)

# Funo para encontrar o caminho euleriano
def find_eulerian_path(graph):
    try:
        path = nx.eulerian_circuit(graph)
        return list(path)
    except nx.NetworkXError:
        return None

# Verifica se G1  euleriano e encontra o caminho euleriano se existir
is_eulerian_G1 = is_eulerian(G1)
eulerian_path_G1 = find_eulerian_path(G1)

# Verifica se G2  euleriano e encontra o caminho euleriano se existir
is_eulerian_G2 = is_eulerian(G2)
eulerian_path_G2 = find_eulerian_path(G2)

# Plotar G1
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.title(f'G1 Euleriano: {is_eulerian_G1}')
nx.draw(G1, with_labels=True, font_weight='bold')

# Plotar G2
plt.subplot(122)
plt.title(f'G2 Euleriano: {is_eulerian_G2}')
nx.draw(G2, with_labels=True, font_weight='bold')

plt.show()

# Imprimir o caminho euleriano de G1, se existir
if is_eulerian_G1:
    print(f'Caminho Euleriano de G1: {eulerian_path_G1}')

# Imprimir o caminho euleriano de G2, se existir
if is_eulerian_G2:
    print(f'Caminho Euleriano de G2: {eulerian_path_G2}')
```
