```python
import networkx as nx
import matplotlib.pyplot as plt

def plot_graph(G, title):
    pos = nx.spring_layout(G)  # Define uma posio para os vrtices
    nx.draw(G, pos, with_labels=True, node_size=500)
    plt.title(title)
    plt.show()

def is_hamiltonian(G):
    n = len(G.nodes())
    if nx.is_connected(G):
        if n >= 3:
            if all(deg >= n / 2 for node, deg in G.degree()):
                return True
    return False

# Definir os grafos
G1 = nx.Graph()
G1.add_edges_from([(1, 2), (1, 4), (2, 3), (2, 4), (3, 4)])
title_G1 = "G1 - Grafo Hamiltoniano"

G2 = nx.Graph()
G2.add_edges_from([(1, 2), (1, 4), (2, 3), (2, 4), (2, 5), (3, 4), (4, 5)])
title_G2 = "G2 - Grafo no Hamiltoniano"

G3 = nx.Graph()
G3.add_edges_from([(1, 2), (1, 4), (1, 5), (1, 5), (2, 3), (2, 4), (2, 5), (3, 4), (4, 5)])
title_G3 = "G3 - Grafo Hamiltoniano"

# Plotar os grafos
plot_graph(G1, title_G1)
plot_graph(G2, title_G2)
plot_graph(G3, title_G3)

# Verificar se os grafos so Hamiltonianos
hamiltonian_G1 = is_hamiltonian(G1)
hamiltonian_G2 = is_hamiltonian(G2)
hamiltonian_G3 = is_hamiltonian(G3)

print(f"{title_G1}: {'Sim' if hamiltonian_G1 else 'No'}  Hamiltoniano")
print(f"{title_G2}: {'Sim' if hamiltonian_G2 else 'No'}  Hamiltoniano")
print(f"{title_G3}: {'Sim' if hamiltonian_G3 else 'No'}  Hamiltoniano")
```
