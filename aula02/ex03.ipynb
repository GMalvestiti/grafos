```python
import networkx as nx
import matplotlib.pyplot as plt

# Definindo as arestas do grafo original G
arestas_G = [(1, 2), (1, 5), (2, 3), (2, 4), (3, 4), (3, 6), (4, 5), (4, 6), (5, 6)]

# Definindo o conjunto de vértices para o subgrafo G1
vertices_G1 = [4, 5, 6]

# Filtrando as arestas que pertencem ao subgrafo G1
arestas_G1 = [aresta for aresta in arestas_G if aresta[0] in vertices_G1 and aresta[1] in vertices_G1]

# Construindo o subgrafo G1
subgrafo_G1 = nx.Graph()
subgrafo_G1.add_nodes_from(vertices_G1)
subgrafo_G1.add_edges_from(arestas_G1)

# Plotando o subgrafo G1
pos = nx.spring_layout(subgrafo_G1)  # Layout do plot
nx.draw(subgrafo_G1, pos, with_labels=True, node_color='lightblue', font_weight='bold', font_size=10, node_size=1000)
plt.title("Subgrafo Induzido G1")
plt.show()
```
