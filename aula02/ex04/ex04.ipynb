```python
import networkx as nx
import matplotlib.pyplot as plt

# Criando o grafo G
G = nx.Graph()
edges = [("1", "2", {"label": "a"}), ("1", "5", {"label": "d"}), ("2", "3", {"label": "b"}),
         ("2", "4", {"label": "c"}), ("3", "4", {"label": "e"}), ("3", "6", {"label": "h"}),
         ("4", "5", {"label": "f"}), ("4", "6", {"label": "g"}), ("5", "6", {"label": "i"})]
G.add_edges_from(edges)

# Lista de arestas do subgrafo induzido G1
edges_g1 = [("2", "4"), ("3", "6"), ("4", "6")]

# Criando o subgrafo induzido G1
G1 = G.edge_subgraph(edges_g1)

# Plotando o grafo G1
pos = nx.spring_layout(G1)  # Definindo o layout do plot
nx.draw(G1, pos, with_labels=True, node_size=1000, font_size=10, font_color='black', font_weight='bold', node_color='lightblue', edge_color='gray')
nx.draw_networkx_edge_labels(G1, pos, edge_labels={(u, v): G[u][v]['label'] for u, v in G1.edges()}, font_color='black')
plt.title("Subgrafo Induzido G1 pelas Arestas c, g e h")
plt.show()
```
