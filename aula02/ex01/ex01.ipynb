```python
import networkx as nx
import matplotlib.pyplot as plt

# Criando o grafo
G = nx.DiGraph()

# Adicionando arestas
edges = [('Alberto', 'Beltrano'), ('Ciclano', 'Alberto'), ('Beltrano', 'Fulana'),
         ('Fulana', 'Deltrana'), ('Beltrano', 'Deltrana'), ('Deltrana', 'Ciclano'),
         ('Gustavo', 'Alberto')]
G.add_edges_from(edges)

# Plot do grafo
pos = nx.spring_layout(G, seed=42)  # Define a posição dos nós
nx.draw(G, pos, with_labels=True, node_size=1000, node_color='skyblue', font_size=10, font_color='black')
plt.title("Grafo Direcionado de Empréstimos Entre Amigos")
plt.show()
```
