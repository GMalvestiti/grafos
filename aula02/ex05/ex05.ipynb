```python
import networkx as nx
import matplotlib.pyplot as plt

# Definindo as arestas do grafo original G
arestas_G = [(1, 2), (1, 3), (2, 4), (3, 4), (3, 5), (4, 6), (4, 7), (6, 8), (6, 9), (7, 9), (7, 10), (8, 10)]

# Definindo o conjunto de vértices para o subgrafo G1
vertices_G1 = [3, 5, 6, 8, 9]

# Filtrando as arestas que pertencem ao subgrafo G1
arestas_G1 = [aresta for aresta in arestas_G if aresta[0] in vertices_G1 and aresta[1] in vertices_G1]

# Construindo o subgrafo G1
subgrafo_G1 = nx.Graph()
subgrafo_G1.add_nodes_from(vertices_G1)
subgrafo_G1.add_edges_from(arestas_G1)

# Plotando o subgrafo G1
pos = nx.spring_layout(subgrafo_G1)  # Layout do plot
nx.draw(subgrafo_G1, pos, with_labels=True, node_color='lightblue', font_weight='bold', font_size=10, node_size=1000)
plt.title("Subgrafo Induzido G1")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
arestas_G = [(1, 2), (1, 3), (2, 4), (3, 4), (3, 5), (4, 6), (4, 7), (6, 8), (6, 9), (7, 9), (7, 10), (8, 10)]
# Criando o grafo G
G = nx.Graph()
edges = [("1", "2", {"label": "a"}), ("1", "3", {"label": "b"}), ("2", "4", {"label": "c"}), ("3", "4", {"label": "d"}),
         ("3", "5", {"label": "e"}), ("4", "6", {"label": "f"}), ("4", "7", {"label": "g"}), ("6", "8", {"label": "h"}),
         ("6", "9", {"label": "i"}), ("7", "9", {"label": "j"}), ("7", "10", {"label": "k"}), ("8", "10", {"label": "l"})]
G.add_edges_from(edges)

# Lista de arestas do subgrafo induzido G1
edges_g1 = [("1", "3"), ("4", "7"), ("7", "9"), ("8", "10")]

# Criando o subgrafo induzido G1
G1 = G.edge_subgraph(edges_g1)

# Plotando o grafo G1
pos = nx.spring_layout(G1)  # Definindo o layout do plot
nx.draw(G1, pos, with_labels=True, node_size=1000, font_size=10, font_color='black', font_weight='bold', node_color='lightblue', edge_color='gray')
nx.draw_networkx_edge_labels(G1, pos, edge_labels={(u, v): G[u][v]['label'] for u, v in G1.edges()}, font_color='black')
plt.title("Subgrafo Induzido G1")
plt.show()
```
